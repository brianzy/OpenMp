```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("time.csv",header=None)
df.index+=1
df[0]=df[0].apply(lambda x: x/27.187368)
df
```

```python
ax=sns.lineplot(x=df.index,y=df[0])
ax.set(xlabel='Number of CPU', ylabel='Speedup')
ax.set_title("OpenMp package parallel computing improvement")
plt.savefig('time.png')

plt.show()
```
